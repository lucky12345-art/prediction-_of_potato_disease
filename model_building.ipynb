```python
len(dataset)
train_size=0.8
len(dataset)*train_size
```

```python
train_ds=dataset.take(54)
```

```python
test_ds=dataset.skip(54)
len(test_ds)
```

```python
validation_size=0.1
len(dataset)*validation_size
```

```python
val_ds=test_ds.take(6)
len(val_ds)
```

```python
test_ds=test_ds.skip(6)
len(test_ds)
```

```python
train_ds=train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds=val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds=test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize_and_rescale=tf.keras.Sequential([
    layers.Resizing(256,256),
    layers.Rescaling(1.0/255)
])
```

```python
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.1)
])
```

```python
model=tf.keras.Sequential([
    resize_and_rescale,
    data_augmentation,
    layers.Conv2D(32,3,padding='same',activation='relu',input_shape=(32,256,256,3)),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(3,activation="softmax")
])

model.build(input_shape=(32,256,256,3))
    
    
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

```python
history=model.fit(
    train_ds,
    epochs=20,
    batch_size=32,
    verbose=1,
    validation_data=val_ds
)
```

```python
model.evaluate(test_ds)
```

```python
for image_batch,label_batch in dataset.take(1):
    plt.imshow(image_batch[0].numpy().astype('uint8'))
    print(f"actual:{class_names[label_batch[0]]}")
      # Expand dims to match model input shape
    predicted = model.predict(tf.expand_dims(image_batch[0], axis=0))
    
    # Get predicted class index
    predicted_class = tf.argmax(predicted[0]).numpy()
    print(f"Predicted: {class_names[predicted_class]}")
    
        
        
    
```
